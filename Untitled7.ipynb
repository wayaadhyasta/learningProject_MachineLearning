3. Preprocessing Teks

```python
df_copy['Ulasan'] = df_copy['Ulasan'].str.lower()
df_copy.head()
```

C1. Normalisasi
Mengubah kata untuk memenuhi tujuan tersendiri
contoh ex. dgn: dengan yg: yang good: bagus

```python
normalisasi = {'dgn': 'dengan', 'gue':'saya', 'bgmn':'bagaimana', 'tdk': 'tidak', 'blum': 'belum', 'gx' : "tidak"}
def normalisasi_teks(str_text):
  for i in normalisasi :
    str_text = str_text.replace(i, normalisasi[i])
  return str_text

df_copy['Ulasan'] = df_copy['Ulasan'].apply(lambda x: normalisasi_teks(x))
df_copy.head()
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, ArrayDictionary

more_stop_word = []

stop_words = StopWordRemoverFactory().get_stop_words()
# Get the default stop words

# Create a StopWordRemover object directly using the default factory
stopword_remover = StopWordRemoverFactory().create_stop_word_remover()

def stopwords(str_text):
  str_text = stopword_remover.remove(str_text)  # Use the remover to remove stop words
  return str_text

df_copy['Ulasan'] = df_copy['Ulasan'].apply(lambda x: stopwords(x))
df_copy.head()
```

C3. Tokenize
Memisahkan kalimat, paragraf atau dokumen menjadi kata-kata atau bagian-bagian tertentu
contoh :
ex. produk ini bagus : ['produl','ini','bagus']  

```python
tokenized = df_copy['Ulasan'].apply(lambda x:x.split())
tokenized
```

Stemming
Digunakan untuk memperkecil jumlah indeks yg berada dai satu data sehingga sebuah kata yang memikliki suffix maupun prefix akan kembali ke bentuk dasarny
contoh :
ex. pesanan -> pesan
membeli -> beli

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
def stemming(Ulasan) :
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()
  dy = []
  for w in Ulasan :
    dt = stemmer.stem(w)
    dy.append(dt)
  d_clean = []
  d_clean = " ".join(dy)
  print(d_clean)
  return d_clean

tokenized = tokenized.apply(stemming)
tokenized.to_csv('cleandata2.csv', index=False)
data_clean = pd.read_csv('cleandata2.csv', encoding='latin1')
data_clean.head()
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV

cv_method = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=999)

params_NB = {'var_smoothing': np.logspace(0, -9, num=100)}
gacv_ab = GridSearchCV(estimator=nb, param_grid=params_NB, cv=cv_method, verbose=1, scoring='accuracy')
gacv_ab.fit(x_train, y_train)
gacv_ab.best_params_
```
